## Import the Mbeere Embu dataset
Let's see the structure of the dataset

```python
import pandas as pd
df = pd.read_excel("Mbeere South.xlsx")
df.head()
```

## Checking the rows and columns in the dataset
we have 17 columns with 33915 rows of data

```python
df.shape
```

## Check the columns available in the dataset

```python
df.columns
```

## Drop the unecessary columns in the database

```python
df.drop([1714514, 'M',14,65,325,74,2,22935462,'EMBU','MBEERE SOUTH','MBETI SOUTH',df.columns[5],df.columns[15]], axis=1,inplace = True)
df.head()
```

## Rename the remaining columns into column names that make sense

```python
df.rename(columns={'NGONDI':'Surname','KANYARIRI PRIMARY SCHOOL':'Account Name',702502701:'Telephone','THOMAS KINYUA':'Names'},inplace = True)
df.head()
```

## Fix the casing of the names

```python
df['Surname'] = df['Surname'].str.title()
df['Names'] = df['Names'].str.title()
df.head()
```

##  Split the name column into Firstname and lastname

```python
df['First Name'], df['Middle Name'] = df['Names'].str.split(' ', 1).str
df
```

## Add the missing columns into the dataframe

```python
df["Job Title"] = "Farmer"
df["Email Address"] = ""
df["Address"] = ""
df["postal code"] = ""
df["city"] = ""
df["Nationality"] = "Kenya"
df["Role"] = "Member"
df["onboardingstatus_c"] = "Registered"

df.head()
```

## Arrange the columns of the dataset into the appropriate order

```python
df = df[['First Name', 'Middle Name', 'Surname', 'Job Title', 'Email Address',
       'Telephone', 'Address', 'postal code', 'city', 'Nationality', 'Role',
       'onboardingstatus_c', 'Account Name']]
df.tail()
```
