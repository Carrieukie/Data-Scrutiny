## Fix the casing of the surname and the Names column

```python
total["Surname"] = total["Surname"].str.title()
total["Names"] = total["Names"].str.title()
total.head()
```

## Split the names column into First name and Middle name

```python
total['First Name'], total['Middle Name'] = total['Names'].str.split(' ', 1).str
total.head()
```

## Drop the unnecesary columns

```python
total.drop(['Names','Ward','KANGEMA','County','Unnamed: 0'], axis = 1, inplace = True)
total.head()
```

## Re arrange the columns

```python
total = total[['First Name','Middle Name','Surname','PhoneNumber','Factory']]
total.head()
```

## See the column names which contain null values

```python
total[total["Surname"].isnull()] 
```

## Fill in the null values of the Surname using values contained in the middle name

```python
total['Surname']=total['Surname'].fillna(total['Middle Name'])
```
